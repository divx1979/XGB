```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score
import xgboost as xgb
```

```python
da1 = pd.read_csv('/Users/divyeshbhatt/Downloads/Datas/Kaggle_Contests/Classification/CreditCard/creditcard.csv')
```

```python
X = da1.drop('Class', axis = 1)
y = da1['Class']
```

```python
## split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
## Handle Imbalanced Data
scale_pos_weight = sum(y_train == 0) / sum(y_train == 1)
```

```python
## Adv. 
mod1 = xgb.XGBClassifier(scale_pos_weight = scale_pos_weight, objective = 'binary:logistic')
```

```python
## Hyperparameter

params = {
    'max_depth' : [3, 5, 7],
    'min_child_weight' : [1, 3, 5],
    'eta' : [0.01, 0.1, 0.3]
}

clf = GridSearchCV(mod1, param_grid = params, scoring = 'roc_auc', cv = 3)
clf.fit(X_train, y_train, early_stopping_rounds = 10, eval_set = [(X_test, y_test)], verbose = True)
```

```python
## Eval Set

best_mod1 = clf.best_estimator_
yhat1 = best_mod1.predict(X_test)
acc1 = accuracy_score(y_test, yhat1)
roc1 = roc_auc_score(y_test, yhat1)
print(acc1)
print(roc1)
```
